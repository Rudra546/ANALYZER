```python
import numpy as np

class NumPyAnalyzer:

    def __init__(self):
        self.arr = None

    @classmethod
    def info(cls):
        print("========== NUMPY ANALYZER ==========")

    def create_array(self):
        print("\nArray Creation:")
        print("1. 1D Array")
        print("2. 2D Array")
        print("3. 3D Array")
        ch = int(input("Enter your choice: "))

        if ch == 1:
            n = int(input("Enter number of elements: "))
            data = list(map(int, input(f"Enter {n} elements: ").split()))
            self.arr = np.array(data)

        elif ch == 2:
            r = int(input("Enter rows: "))
            c = int(input("Enter columns: "))
            data = list(map(int, input(f"Enter {r*c} elements: ").split()))
            self.arr = np.array(data).reshape(r, c)

        elif ch == 3:
            d = int(input("Enter depth: "))
            r = int(input("Enter rows: "))
            c = int(input("Enter columns: "))
            data = list(map(int, input(f"Enter {d*r*c} elements: ").split()))
            self.arr = np.array(data).reshape(d, r, c)

        else:
            print("Invalid choice!")
            return

        print("\nArray Created Successfully:")
        print(self.arr)
```

```python
 def indexing_slicing(self):
        if self.arr is None:
            print("Create array first!")
            return

        print("\n1. Indexing")
        print("2. Slicing")
        ch = int(input("Enter choice: "))

        dims = self.arr.ndim

        if ch == 1:
            index = tuple(map(int, input(f"Enter {dims} index values: ").split()))
            print("Element =", self.arr[index])

        elif ch == 2:
            slices = []
            for i in range(dims):
                s = input(f"Dimension {i} slice (start:end): ")
                start, end = map(int, s.split(":"))
                slices.append(slice(start, end))

            print("Sliced Array:")
            print(self.arr[tuple(slices)])

        else:
            print("Invalid choice!")
```

```python
   def combine_split(self):
        if self.arr is None:
            print("Create array first!")
            return

        print("\n1. Combine Arrays")
        print("2. Split Array")
        ch = int(input("Enter choice: "))

        if ch == 1:
            data = list(map(int, input(f"Enter {self.arr.size} elements: ").split()))
            arr2 = np.array(data).reshape(self.arr.shape)
            combined = np.concatenate((self.arr, arr2), axis=0)
            print("Combined Array:")
            print(combined)

        elif ch == 2:
            parts = int(input("Enter number of splits: "))
            split_arr = np.array_split(self.arr, parts)
            print("Split Arrays:")
            for s in split_arr:
                print(s)

   
```

```python
def search_sort_filter(self):
        if self.arr is None:
            print("Create array first!")
            return

        print("\n1. Search Value")
        print("2. Sort Array")
        print("3. Filter Values")
        ch = int(input("Enter choice: "))

        if ch == 1:
            val = int(input("Enter value to search: "))
            print("Found at indices:", np.where(self.arr == val))

        elif ch == 2:
            order = input("Ascending (a) or Descending (d): ")
            flat = np.sort(self.arr, axis=None)
            if order == 'a':
                print(flat)
            else:
                print(flat[::-1])

        elif ch == 3:
            cond = int(input("Show values greater than: "))
            print(self.arr[self.arr > cond])
```

```python
 def aggregates_statistics(self):
        if self.arr is None:
            print("Create array first!")
            return

        print("\n1. Sum")
        print("2. Mean")
        print("3. Median")
        print("4. Standard Deviation")
        print("5. Variance")
        print("6. Min & Max")
        print("7. Percentile")
        ch = int(input("Enter choice: "))

        if ch == 1:
            print("Sum =", np.sum(self.arr))
        elif ch == 2:
            print("Mean =", np.mean(self.arr))
        elif ch == 3:
            print("Median =", np.median(self.arr))
        elif ch == 4:
            print("Std Dev =", np.std(self.arr))
        elif ch == 5:
            print("Variance =", np.var(self.arr))
        elif ch == 6:
            print("Min =", np.min(self.arr), "Max =", np.max(self.arr))
        elif ch == 7:
            p = int(input("Enter percentile: "))
            print(f"{p}th Percentile =", np.percentile(self.arr, p))

```

```python
def main():
    na = NumPyAnalyzer()
    NumPyAnalyzer.info()

    while True:
        print("\n===== MAIN MENU =====")
        print("1. Create Array")
        print("2. Indexing & Slicing")
        print("3. Mathematical Operations")
        print("4. Combine And Split Arrays")
        print("5. Search,Sort and Filter Array")
        print("6. Aggregate & Statistics")
        print("7. Exit")

        choice = int(input("Enter your choice: "))

        if choice == 1:
            na.create_array()
        elif choice == 2:
            na.indexing_slicing()
        elif choice == 3:
            na.mathematical_operations()
        elif choice == 4:
            na.combine_split()
        elif choice == 5:
            na.search_sort_filter()
        elif choice == 6:
            na.aggregates_statistics()
        elif choice == 7:
            print("Program End. Thank You!")
            break
        else:
            print("Invalid choice!")

main()
```
